```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.manifold import Isomap
from sklearn.manifold import TSNE

%matplotlib inline
```

```python
wine=pd.read_csv('Wine.csv');
wine.head()
```

```python
wine.info()
```

```python
wine['Customer_Segment'].unique()
```

```python
X=wine.drop('Customer_Segment', axis=1)
y=wine['Customer_Segment']
```

```python
wine_corr=X.corr()
plt.figure(figsize=(20, 20))
sns.heatmap(wine_corr, annot=True, cmap='RdBu_r')
```

# Based on the heat map, falvonoids, Total_phenols have linear relations
# falavoniods, OD280 ha

```python
sns.scatterplot(x='Flavanoids', y='Total_Phenols', data=wine)
```

```python
sns.scatterplot(x='Flavanoids', y='OD280', data=wine)
```

# Dimensionality Reduction
1. PCA

```python
pca=PCA(n_components=2)
wine_pca=pca.fit_transform(X)
print("Explained variance per principle component: {}".format(pca.explained_variance_ratio_))
```

```python
wine_pca_df=pd.DataFrame(data=wine_pca, columns=['Principle Component 1', 'Principle Component 2'])
wine_pca_df.head()
```

```python
plt.figure(figsize=(10, 8))
plt.scatter(wine_pca_df['Principle Component 1'], wine_pca_df['Principle Component 2'], c=y)
```

2. Isomap technique

```python
embedding=Isomap(n_components=2, n_neighbors=40)
X_isomap=embedding.fit_transform(X)
plt.figure(figsize=(10, 8))
plt.scatter(X_isomap[:,0], X_isomap[:, 1], c=y)
```

3. t-SNE method

```python
tsneModel = TSNE(n_components=2, random_state=0, perplexity=50, n_iter=5000)
tsne_X =  tsneModel.fit_transform(X)

tsne_X = np.vstack((tsne_X.T, y)).T

tsne_wine_df = pd.DataFrame(data=tsne_X, columns=('Dim_1', 'Dim_2', 'label'))

sns.FacetGrid(tsne_wine_df, hue='label', height=6).map(plt.scatter,  'Dim_1', 'Dim_2')
plt.show();
```

# Standarization

```python
from sklearn import preprocessing
```

```python
minmax=preprocessing.MinMaxScaler()
scaler_df=minmax.fit_transform(X)
scaler_df=pd.DataFrame(data=scaler_df, columns=X.columns)
scaler_df.head()
```

```python
fig, (ax1, ax2)=plt.subplots(ncols=2, figsize=(6, 5))
ax1.set_title('Before Scaling')
sns.kdeplot(X['Alcohol'], ax=ax1)
sns.kdeplot(X['Magnesium'], ax=ax1)
sns.kdeplot(X['Proline'], ax=ax1)

ax2.set_title('After Scaling')
sns.kdeplot(scaler_df['Alcohol'], ax=ax2)
sns.kdeplot(scaler_df['Magnesium'], ax=ax2)
sns.kdeplot(scaler_df['Proline'], ax=ax2)

plt.show()
```

# Clustering Techniques (K-Means)
Finding the no. of clusters: i)Silhouette method

```python
from sklearn import metrics
from sklearn.metrics import silhouette_score
from sklearn.cluster import KMeans
```

```python
sScore=[]
for n_cluster in range(2, 30):
    kmeans=KMeans(init='k-means++', n_clusters=n_cluster, n_init=30)
    kmeans.fit(scaler_df)
    clusters=kmeans.predict(scaler_df)
    sScore_avg=silhouette_score(scaler_df, clusters)
    sScore.append(sScore_avg)
silhouette_df=pd.DataFrame({'No_clusters':range(2, 30), 'Silhouette_score':sScore})
silhouette_df.head()
```

```python
sns.set()
plt.figure(figsize=(10, 6))
ax=sns.lineplot(x='No_clusters', y='Silhouette_score', data=silhouette_df)
ax.set(xticks=silhouette_df['No_clusters'])
plt.show()
```

```python
#Highest silhouette score occurs at no_clusters=3, no. of clusters required for the given dataset=3
```

# Visualising K-Means (K=3)

```python
kmeans_tsne_df=tsne_wine_df.drop('label', axis=1)
kmeans_tsne=KMeans(n_clusters=3, random_state=0)
kmeans_tsne_labels=kmeans_tsne.fit_predict(kmeans_tsne_df)
kmeans_tsne_centers=kmeans_tsne.cluster_centers_
kmeans_tsne_df['label']=kmeans_tsne_labels.tolist()
```

```python
kmeans_tsne_df.head()
```

```python
from scipy.spatial.distance import cdist

fig, ax = plt.subplots(figsize=(10, 10))
ax.axis('equal')
sns.scatterplot(x='Dim_1', y='Dim_2', hue='label', data=kmeans_tsne_df, palette=['red','blue','green'], ax=ax)
plt.scatter(kmeans_tsne_centers[:,0], kmeans_tsne_centers[:, 1], s = 300, c = 'black' , label = 'centeroid', marker='*')
radii = [cdist(kmeans_tsne_df[kmeans_tsne_labels == i].drop('label', axis=1), [center]).max() for i, center in enumerate(kmeans_tsne_centers)]
for c, r in zip(kmeans_tsne_centers, radii):
    ax.add_patch(plt.Circle(c, r, alpha=0.3))
```

# Guassian mixture model

```python
from sklearn.mixture import GaussianMixture
gmm_tsne_df=tsne_wine_df.drop('label', axis=1)
gmm=GaussianMixture(n_components=3, random_state=42, covariance_type='full')
gmm_tsne_labels=gmm.fit_predict(gmm_tsne_df)
gmm_tsne_centers=gmm.means_
gmm_tsne_df['label']=gmm_tsne_labels.tolist()
gmm_tsne_df.head()
```

```python
from matplotlib.patches import Ellipse


fig, ax = plt.subplots(figsize=(10, 10))
ax.axis('equal')
sns.scatterplot(x='Dim_1', y='Dim_2', hue='label', data=gmm_tsne_df, palette=['red','blue','green'], ax=ax)

plt.scatter(gmm_tsne_centers[:,0], gmm_tsne_centers[:, 1], s = 300, c = 'black' , label = 'centeroid', marker='*')
ax.add_patch(Ellipse(gmm_tsne_centers[0], 2, 5.5, 150, alpha=0.3));
ax.add_patch(Ellipse(gmm_tsne_centers[1], 2, 6.6, 145, alpha=0.3));
ax.add_patch(Ellipse(gmm_tsne_centers[2], 2, 7, 145, alpha=0.3));

plt.show()
```
